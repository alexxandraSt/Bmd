```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method

plt.plot(x, y) # build plot
plt.xlabel('X label') # add name of x axis
plt.ylabel('Y label') # add name of y axis
plt.title('Title') # add title
```

```python
# create multiply plots
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')

axes2.plot(x, y, 'b')
axes2.set_xlabel('Y label')
axes2.set_ylabel('X label')
axes2.set_title('Title')
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig,axes = plt.subplots(nrows=3, ncols=4)
plt.tight_layout()

axes[0,1].plot(x, y)
axes[1,0].plot(y, x)
axes[2,2].plot(x, y)
axes[2,3].plot(y, x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)

fig,ax = plt.subplots()
ax.plot(x, x**2, color = 'g', lw = 2, ls = '--')
ax.plot(x, x**3, color = 'y', marker = 'o', markersize = 10, markerfacecolor = 'b')

fig.savefig("lab2.jpg")
```
